```python
import os
import pandas as pd
```

```python
daily=pd.read_csv('daily_sales.csv')
monthly=pd.read_csv('monthly_sales.csv')
daily=daily.rename(columns={'Unnamed: 0':'date'})
```

```python
daily_HM=daily[[s for s in daily.columns if "HM" in s]]
daily_HM['date']=daily['date'].values
HM_date=daily_HM.groupby('date').mean()
HM_date.to_csv('HM_date.csv')
```

```python
daily_HM=daily[[s for s in daily.columns if "CF" in s]]
daily_HM['date']=daily['date'].values
CF_date=daily_HM.groupby('date').mean()
CF_date.to_csv('CF_date.csv')
```

```python
daily_HM=daily[[s for s in daily.columns if "FF" in s]]
daily_HM['date']=daily['date'].values
FF_date=daily_HM.groupby('date').mean()
FF_date.to_csv('FF_date.csv')
```
